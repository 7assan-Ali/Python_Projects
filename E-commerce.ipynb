```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('ecommerce_customer_behavior_dataset_v2.csv')
df
```

```python
df=df.drop(['Order_ID','Customer_ID','Date',],axis=1)
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
best_Category=df.groupby('Product_Category').size()
best_Category=best_Category.sort_values(ascending=True)
best_Category
```

```python
best_Category.plot(
    kind='pie',legend=True,autopct='%1.2f%%',figsize=(12,9)
)
```

```python
best_device=df.groupby('Device_Type').size()
best_device=best_device.sort_values(ascending=True)
best_device
```

```python
best_device.plot(
                 kind='pie',legend=True,autopct='%1.2f%%',figsize=(12,9)
                )
```

```python
City=df.groupby('City').size()
City=City.sort_values(ascending=False)
City
```

```python
C=City.plot(
    kind='bar',
    rot=45,
    xlabel='value',
    ylabel='city',
    title='City Bar Chart',
    edgecolor='black'
)
for i, v in enumerate(City):   # أو أي Series فيها القيم
    C.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=10)
```
