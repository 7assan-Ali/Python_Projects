```python
print(plt.style.available)
plt.style.use("tableau-colorblind10")
```

```python
groupby_company_size=df.groupby("company_size")["salary"].mean()
groupby_company_size
```

```python
groupby_company_size.plot(kind='bar',xlabel='company_size',ylabel='Average of salary')
```

```python
groupby_company_size.plot(kind='barh',xlabel='company_size',ylabel='Average of salary')
```

```python
groupby_company_size.plot(kind='line',xlabel='company_size',ylabel='Average of salary')
```

```python
groupby_company_size.plot(
    kind='pie',
    legend=True,
    figsize=(7,7),
    autopct='%1.2f%%',title='Average of salary')
```

```python
groupby_company_size.plot.area(
    color='green',        # ← اللون اللي تختاره
    figsize=(8,6),
    title='Average Salary by Company Size'
)
```
