```python
groupby_Net_Banking=df.groupby("Payment_Method")["Discount (%)"].mean()
groupby_Net_Banking
```

```python
groupby_Net_Banking.plot(kind='bar',xlabel='payment_method',ylabel='AMount_of_discount')
```

```python
groupby_Net_Banking.plot(kind='line',xlabel='payment_method',ylabel='AMount_of_discount')
```
